```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2' 
```

```python
import tensorflow as tf
```

```python
efficientnet_open_waste_model = tf.keras.models.load_model('models/efficientnet_v2_open_waste.keras')
```

```python
# Convert it to SavedModel format:
efficientnet_open_waste_model.export('models/efficientnet_v2_open_waste')
```

```python
# Convert it to TFLite
converter = tf.lite.TFLiteConverter.from_saved_model('models/efficientnet_v2_open_waste')
efficientnet_open_waste_model_tflite = converter.convert()
```

```python
with open('models/efficientnet_v2_open_waste.tflite', 'wb') as f:
    f.write(efficientnet_open_waste_model_tflite)
```
